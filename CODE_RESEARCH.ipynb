```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s\.]', '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
    preprocessed_text = ' '.join(lemmatized_tokens)
    return preprocessed_text

# Apply preprocessing
ml_text = preprocess_text(ml_data['text'])
st_text = preprocess_text(stock_market_data['text'])
bl_text = preprocess_text(blockchain_data['text'])

# Display preprocessed text
print(ml_text[:500])
print(st_text[:500])
print(bl_text[:500])
```

```python
def analyze_terms(extracted_terms, true_terms):
    extracted_terms = {term.lower().strip() for term in extracted_terms}
    true_terms = {term.lower().strip() for term in true_terms}

    TP = extracted_terms & true_terms
    FP = extracted_terms - true_terms
    FN = true_terms - extracted_terms

    precision = len(TP) / (len(TP) + len(FP)) if (len(TP) + len(FP)) > 0 else 0.0
    recall = len(TP) / (len(TP) + len(FN)) if (len(TP) + len(FN)) > 0 else 0.0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0


    all_terms = true_terms | extracted_terms
    true_labels = np.array([1 if term in true_terms else 0 for term in all_terms])
    pred_labels = np.array([1 if term in extracted_terms else 0 for term in all_terms])


    rmse = np.sqrt(np.mean((true_labels - pred_labels) ** 2))

    return {
        "True Positives (TP)": TP,
        "False Positives (FP)": FP,
        "False Negatives (FN)": FN,
        "Precision": round(precision, 2),
        "Recall": round(recall, 2),
        "F1-score": round(f1_score, 2),
        "RMSE": round(rmse, 2)
    }
```
